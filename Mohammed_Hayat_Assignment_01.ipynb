# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
import numpy as np

print("r    V")
for r in range(4, 20, 4):  # Loop through even numbers from 2 to 20
    V = (2/3) * np.pi * (r ** 3)  # Compute hemisphere volume
    print(f"{r:<3} {V:.2f}")  # Print r and V formatted to 2 decimal places


```

# Question 3 (10 points).

Create a Python program that encompasses various string manipulation task:

(a) Prompt the user to input two sentences.

(b) Transform each sentence into a list, where each element corresponds to a word in the sentence. For example, if the sentences are "Hello, how are you doing today?" and "This is an example sentence for the program!"

The program should output:
['Hello', ',', 'how', 'are', 'you', 'doing', 'today', '?', 'This', 'is', 'an', 'example', 'sentence', 'for', 'the', 'program', '!']

(c) Compare the two sentences. Print out a list of words that occur in both sentences.

(d) Remove any punctuation from the sentence and print the sanitized sentences. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”).

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python
import string

# (a) Prompt the user to input two sentences
sentence1_input = input("Enter the first sentence: ")
sentence2_input = input("Enter the second sentence: ")

# (b) Split each sentence into a list of words, keeping punctuation as separate elements
def split_into_words(sentence):
    words = []
    current_word = ""
    for char in sentence:
        if char in string.punctuation or char == " ":
            if current_word:
                words.append(current_word)
                current_word = ""
            if char != " ":
                words.append(char)
        else:
            current_word += char
    if current_word:
        words.append(current_word)
    return words

words_in_sentence1 = split_into_words(sentence1_input)
words_in_sentence2 = split_into_words(sentence2_input)

# (c) Compare the two sentences and find common words
common_words = [word for word in words_in_sentence1 if word in words_in_sentence2]

# (d) Remove punctuation from each sentence and print the cleaned sentences
def remove_punctuation(text):
    return ''.join([char for char in text if char not in string.punctuation])

cleaned_sentence1 = remove_punctuation(sentence1_input)
cleaned_sentence2 = remove_punctuation(sentence2_input)

# (e) Count the occurrences of vowels (a, e, i, o, u) in each sentence
def count_vowels(sentence):
    vowels = "aeiou"
    vowel_count = {vowel: 0 for vowel in vowels}
    for char in sentence.lower():
        if char in vowels:
            vowel_count[char] += 1
    return vowel_count

vowels_in_sentence1 = count_vowels(cleaned_sentence1)
vowels_in_sentence2 = count_vowels(cleaned_sentence2)

# Print results
print(f"\nWords in the first sentence: {words_in_sentence1}")
print(f"Words in the second sentence: {words_in_sentence2}")
print(f"\nCommon words in both sentences: {common_words}")
print(f"\nCleaned sentence 1 (no punctuation): {cleaned_sentence1}")
print(f"Cleaned sentence 2 (no punctuation): {cleaned_sentence2}")
print(f"\nVowel counts in sentence 1: {vowels_in_sentence1}")
print(f"Vowel counts in sentence 2: {vowels_in_sentence2}")
```

# Question 4 (10 Points)

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(c) Enhance your calculator to include additional mathematical operations such as exponentiation, square root, and modulus.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
import math

# (a) Function to evaluate basic expressions (addition, subtraction, multiplication, division)
def evaluate_expression(expression):
    try:
        result = eval(expression)
        return result
    except ZeroDivisionError:
        return "Error: Division by zero is not allowed."
    except Exception as e:
        return f"Error: Invalid expression - {e}"

# (b) Function to handle parentheses by using Python's eval function, which handles nested parentheses
def handle_parentheses():
    print("Simple Calculator")
    print("Supported operations: +, -, *, /, ** (exponentiation), sqrt (square root), % (modulus), parentheses for grouping")
    print("Type 'exit' to quit.")

    while True:
        user_input = input("Enter an expression: ").strip()

        if user_input.lower() == 'exit':
            print("Exiting the calculator. Goodbye!")
            break

        # (c) Handle additional mathematical operations and evaluate the expression
        if "sqrt" in user_input:
            try:
                number = float(user_input.split('(')[1].split(')')[0])
                result = math.sqrt(number)
                print(f"Result: {result}")
            except ValueError:
                print("Error: Invalid input for square root.")
        else:
            result = evaluate_expression(user_input)
            print(f"Result: {result}")

# (d) Enhanced error handling already implemented in `evaluate_expression` function
# - Handles division by zero
# - Handles invalid expressions (e.g., malformed input)

# (e) Command to exit is implemented in the while loop, where 'exit' breaks the loop
handle_parentheses()
```

# Question 5 (20 points).

In the field of Data Science, data is often formatted as a comma-delimited (CSV) file, where each line in a file corresponds to a field's value. Refer to Chapter 9 (p. 106) for more information on comma-delimited files.

To answer this question, you will need to download the file [Assignment1_denton_housing.csv](https://raw.githubusercontent.com/unt-iialab/info5731-spring2022/main/assignments/Assignment1_denton_housing.csv) included with this assignment. The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----
	2012 ----- -----
    2013 ----- -----
	2014 ----- -----
  
  

(c) (5 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (5 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
import pandas as pd
from statistics import harmonic_mean

# Load the CSV file
# Ensure the file is in the same directory
import pandas as pd

# Corrected way to read the CSV file
df = pd.read_csv("/content/Assignment1_denton_housing (1).csv")


# Clean up column names
df.columns = df.columns.str.strip()

# Convert 'year' and 'value' columns to integer type
df['year'] = df['year'].astype(int)
df['value'] = df['value'].astype(int)

# Restructure the data to make it easier to analyze
df_pivot = df.pivot(index='year', columns='title_field', values='value')

# (a) Calculate Yearly Changes in Occupied Housing Units
yearly_changes = df_pivot['Occupied Housing Units'].diff().dropna().abs()
yearly_changes_sorted = yearly_changes.sort_values()

# Print the results in ascending order
print("Yearly Change in Occupied Housing Units:")
for years, change in yearly_changes_sorted.items():
    print(f"{int(years - 1)}-{int(years)}: {int(change)}")

# (b) Compute and Print Percentages of Vacant and Occupied Housing Units
print("\nYear  Vacant Housing Units  Occupied Housing Units")
for year in df_pivot.index:
    total_units = df_pivot.loc[year, 'Vacant Housing Units'] + df_pivot.loc[year, 'Occupied Housing Units']
    vacant_percent = (df_pivot.loc[year, 'Vacant Housing Units'] / total_units) * 100
    occupied_percent = (df_pivot.loc[year, 'Occupied Housing Units'] / total_units) * 100
    print(f"{year}   {vacant_percent:.2f}%   {occupied_percent:.2f}%")

# (c) Find Year with Highest Vacant and Occupied Housing Units
max_vacant_year = df_pivot['Vacant Housing Units'].idxmax()
max_vacant_value = df_pivot['Vacant Housing Units'].max()

max_occupied_year = df_pivot['Occupied Housing Units'].idxmax()
max_occupied_value = df_pivot['Occupied Housing Units'].max()

print("\nYear with the Highest Housing Units:")
print(f"Vacant Housing Units: {max_vacant_year} with {max_vacant_value} units")
print(f"Occupied Housing Units: {max_occupied_year} with {max_occupied_value} units")

# (d) Compute and Print Harmonic Mean of Total Housing Units
total_units_list = df_pivot['Vacant Housing Units'] + df_pivot['Occupied Housing Units']
harmonic_mean_value = harmonic_mean(total_units_list)

print(f"\nHarmonic Mean of Total Housing Units: {harmonic_mean_value:.2f}")

```

# Question 6 (20 points).
Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

```python
import csv
import os
import random
from collections import defaultdict

# File name for inventory storage
INVENTORY_FILE = "inventory.csv"

# Product categories
CATEGORIES = ["Electronics", "Groceries", "Clothing", "Home Appliances", "Toys"]

# Edge Case Handling: Ensure file exists
if not os.path.exists(INVENTORY_FILE):
    with open(INVENTORY_FILE, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Product ID", "Name", "Category", "Price", "Stock"])

# Function to load inventory
def load_inventory():
    inventory = []
    try:
        with open(INVENTORY_FILE, "r") as file:
            reader = csv.DictReader(file)
            for row in reader:
                if row.get("Product ID") and row.get("Name") and row.get("Category"):
                    inventory.append({
                        "Product ID": row["Product ID"],
                        "Name": row["Name"],
                        "Category": row["Category"],
                        "Price": float(row["Price"]),
                        "Stock": int(row["Stock"])
                    })
    except Exception as e:
        print(f"Error loading inventory: {e}")
    return inventory

# Function to save inventory
def save_inventory(inventory):
    with open(INVENTORY_FILE, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Product ID", "Name", "Category", "Price", "Stock"])
        for item in inventory:
            writer.writerow([item["Product ID"], item["Name"], item["Category"], item["Price"], item["Stock"]])

# Function to add a new product
def add_product():
    inventory = load_inventory()
    product_id = str(random.randint(1000, 9999))  # Generate unique product ID
    name = input("Enter product name: ").strip()
    category = input(f"Enter category ({', '.join(CATEGORIES)}): ").strip()
    if category not in CATEGORIES:
        print("Invalid category. Try again.")
        return
    price = float(input("Enter product price: "))
    stock = int(input("Enter stock quantity: "))

    # Edge Case: Prevent negative price or stock
    if price < 0 or stock < 0:
        print("Price and stock must be non-negative.")
        return

    # Check for duplicate product name
    for item in inventory:
        if item["Name"].lower() == name.lower():
            print("Product already exists.")
            return

    inventory.append({"Product ID": product_id, "Name": name, "Category": category, "Price": price, "Stock": stock})
    save_inventory(inventory)
    print(f"Product '{name}' added successfully!")

# Function to delete a product
def delete_product():
    inventory = load_inventory()
    product_id = input("Enter Product ID to delete: ").strip()

    updated_inventory = [item for item in inventory if item["Product ID"] != product_id]

    if len(updated_inventory) == len(inventory):
        print("Product ID not found.")
        return

    save_inventory(updated_inventory)
    print("Product deleted successfully!")

# Function to update product details
def update_product():
    inventory = load_inventory()
    product_id = input("Enter Product ID to update: ").strip()

    for item in inventory:
        if item["Product ID"] == product_id:
            item["Name"] = input("Enter new name: ").strip() or item["Name"]
            item["Category"] = input(f"Enter new category ({', '.join(CATEGORIES)}): ").strip() or item["Category"]
            item["Price"] = float(input("Enter new price: ") or item["Price"])
            item["Stock"] = int(input("Enter new stock quantity: ") or item["Stock"])
            save_inventory(inventory)
            print("Product updated successfully!")
            return

    print("Product ID not found.")

# Function to search for products
def search_product():
    inventory = load_inventory()
    keyword = input("Enter product name or ID to search: ").strip().lower()

    results = [item for item in inventory if keyword in item["Name"].lower() or keyword in item["Product ID"]]

    if results:
        print("\nSearch Results:")
        for item in results:
            print(f"{item['Product ID']} | {item['Name']} | {item['Category']} | ${item['Price']} | Stock: {item['Stock']}")
    else:
        print("No products found.")

# Function to automate restocking
def restock():
    inventory = load_inventory()
    for item in inventory:
        if item["Stock"] < 10:  # Restock threshold
            item["Stock"] += 20  # Restock by adding 20 units
    save_inventory(inventory)
    print("Restocking completed.")

# Function to generate reports
def generate_report():
    inventory = load_inventory()
    category_summary = defaultdict(float)

    print("\nInventory Report:")
    for item in inventory:
        print(f"{item['Product ID']} | {item['Name']} | {item['Category']} | ${item['Price']} | Stock: {item['Stock']}")
        category_summary[item["Category"]] += item["Price"] * item["Stock"]

    print("\nCategory Breakdown:")
    for category, value in category_summary.items():
        print(f"{category}: ${value:.2f}")

# Main menu
def main():
    while True:
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Update Product")
        print("4. Search Product")
        print("5. Restock Items")
        print("6. Generate Report")
        print("7. Exit")

        choice = input("Enter choice: ")
        if choice == "1":
            add_product()
        elif choice == "2":
            delete_product()
        elif choice == "3":
            update_product()
        elif choice == "4":
            search_product()
        elif choice == "5":
            restock()
        elif choice == "6":
            generate_report()
        elif choice == "7":
            print("Exiting system...")
            break
        else:
            print("Invalid choice. Try again.")

# Run the program
if __name__ == "__main__":
    main()
```

# Question 7 (20 points)

Create a well desined Python program for a flight booking system. Offer core functions for adding new flights, finding trips, reserving and releasing seating space and creating performance reports. Ensure the system manages edge cases like invalid input, overbooking, and cancellations. Finally, extend the booking system by including advanced features and maintain about 50 flight in the system.

```python
import csv
import os
import random
from collections import defaultdict

# File to store flight details
FLIGHT_FILE = "flights.csv"

# Ensuring the file exists before performing operations
if not os.path.exists(FLIGHT_FILE):
    with open(FLIGHT_FILE, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Flight ID", "Airline", "Origin", "Destination", "Date", "Time", "Total Seats", "Available Seats", "Price"])

# Function to load flight data
def load_flights():
    flights = []
    try:
        with open(FLIGHT_FILE, "r") as file:
            reader = csv.DictReader(file)
            for row in reader:
                flights.append({
                    "Flight ID": row["Flight ID"],
                    "Airline": row["Airline"],
                    "Origin": row["Origin"],
                    "Destination": row["Destination"],
                    "Date": row["Date"],
                    "Time": row["Time"],
                    "Total Seats": int(row["Total Seats"]),
                    "Available Seats": int(row["Available Seats"]),
                    "Price": float(row["Price"])
                })
    except Exception as e:
        print(f"Error loading flight data: {e}")
    return flights

# Function to save flight data
def save_flights(flights):
    with open(FLIGHT_FILE, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Flight ID", "Airline", "Origin", "Destination", "Date", "Time", "Total Seats", "Available Seats", "Price"])
        for flight in flights:
            writer.writerow([flight["Flight ID"], flight["Airline"], flight["Origin"], flight["Destination"],
                             flight["Date"], flight["Time"], flight["Total Seats"], flight["Available Seats"], flight["Price"]])

# Function to add a new flight
def add_flight():
    flights = load_flights()
    flight_id = str(random.randint(1000, 9999))  # Generate unique Flight ID
    airline = input("Enter airline name: ").strip()
    origin = input("Enter origin airport: ").strip()
    destination = input("Enter destination airport: ").strip()
    date = input("Enter flight date (YYYY-MM-DD): ").strip()
    time = input("Enter departure time (HH:MM AM/PM): ").strip()
    total_seats = int(input("Enter total number of seats: "))
    price = float(input("Enter ticket price: "))

    if total_seats <= 0 or price < 0:
        print("Invalid input! Seats and price must be positive numbers.")
        return

    flights.append({
        "Flight ID": flight_id,
        "Airline": airline,
        "Origin": origin,
        "Destination": destination,
        "Date": date,
        "Time": time,
        "Total Seats": total_seats,
        "Available Seats": total_seats,
        "Price": price
    })

    save_flights(flights)
    print(f"Flight {flight_id} from {origin} to {destination} added successfully!")

# Function to search for flights
def search_flight():
    flights = load_flights()
    destination = input("Enter destination airport: ").strip().lower()
    date = input("Enter travel date (YYYY-MM-DD): ").strip()

    matching_flights = [flight for flight in flights if flight["Destination"].lower() == destination and flight["Date"] == date]

    if matching_flights:
        print("\nAvailable Flights:")
        for flight in matching_flights:
            print(f"{flight['Flight ID']} | {flight['Airline']} | {flight['Origin']} → {flight['Destination']} | {flight['Date']} {flight['Time']} | ${flight['Price']} | Seats: {flight['Available Seats']}")
    else:
        print("No flights found.")

# Function to book a seat
def book_seat():
    flights = load_flights()
    flight_id = input("Enter Flight ID to book a seat: ").strip()

    for flight in flights:
        if flight["Flight ID"] == flight_id:
            if flight["Available Seats"] > 0:
                flight["Available Seats"] -= 1
                save_flights(flights)
                print(f"Seat booked successfully on Flight {flight_id}. Remaining seats: {flight['Available Seats']}")
            else:
                print("Flight is fully booked!")
            return

    print("Flight ID not found.")

# Function to cancel a booking
def cancel_booking():
    flights = load_flights()
    flight_id = input("Enter Flight ID to cancel booking: ").strip()

    for flight in flights:
        if flight["Flight ID"] == flight_id:
            if flight["Available Seats"] < flight["Total Seats"]:
                flight["Available Seats"] += 1
                save_flights(flights)
                print(f"Booking canceled. Seat released for Flight {flight_id}.")
            else:
                print("No active bookings to cancel.")
            return

    print("Flight ID not found.")

# Function to generate flight reports
def generate_report():
    flights = load_flights()
    destination_count = defaultdict(int)

    print("\nFlight Report:")
    for flight in flights:
        print(f"{flight['Flight ID']} | {flight['Airline']} | {flight['Origin']} → {flight['Destination']} | {flight['Date']} | Seats Available: {flight['Available Seats']}")
        destination_count[flight["Destination"]] += 1

    print("\nPopular Destinations:")
    for destination, count in destination_count.items():
        print(f"{destination}: {count} flights")

# Main menu
def main():
    while True:
        print("\nFlight Booking System")
        print("1. Add Flight")
        print("2. Search Flights")
        print("3. Book Seat")
        print("4. Cancel Booking")
        print("5. Generate Report")
        print("6. Exit")

        choice = input("Enter choice: ")
        if choice == "1":
            add_flight()
        elif choice == "2":
            search_flight()
        elif choice == "3":
            book_seat()
        elif choice == "4":
            cancel_booking()
        elif choice == "5":
            generate_report()
        elif choice == "6":
            print("Exiting system...")
            break
        else:
            print("Invalid choice. Try again.")

# Run the program
if __name__ == "__main__":
    main()

```
